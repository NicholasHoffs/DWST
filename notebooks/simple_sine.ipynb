```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

```python
sample_rate = 44100
step = 1/sample_rate
frequency = 440
```

```python
t = np.arange(0, 3, step)
y = np.sin(2*np.pi*frequency*t)
```

```python
plt.plot(y[0:440])
```

```python
write('simple_440.wav', rate=sample_rate, data=y)
```
