```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

```python
sample_rate = 44100
step = 1/sample_rate
lfo_frequency = 2

t = np.arange(0, 3, step)

modulated_frequency = 440 + 100 * np.sin(2 * np.pi  * lfo_frequency * t)
modulated_phase = np.cumsum(modulated_frequency * step)
```

```python
y = np.sin(2*np.pi*modulated_phase)
```

```python
plt.plot(y[0:440])
```

```python
write('audio/modulated_sine.wav', rate=sample_rate, data=y)
```
